```python
txt_list = glob('./test_pistol_Kalman2/labels/*.txt') #Kalman
print(len(txt_list))
txt_list2 = glob('./test_pistol_yolov5/labels/*.txt')#YOLO
print(len(txt_list2))
txt_list3 = glob('./test_pistol_Kalman3/labels/*.txt')
print(len(txt_list2))
```

```python
with open('./test_pistol_Kalman2/test.txt', 'w') as f:
  f.write('\n'.join(txt_list) + '\n')
```

```python
with open('./test_pistol_yolov5/test2.txt', 'w') as f:
  f.write('\n'.join(txt_list2) + '\n')
```

```python
with open('./test_pistol_Kalman3/test.txt', 'w') as f:
  f.write('\n'.join(txt_list3) + '\n')
```

```python
filenames = []
f = open('./test_pistol_Kalman2/test.txt', 'r')
while True:
    line = f.readline()
    line = line[0:-1]
    filenames.append((line))
    if not line: 
        print(len(filenames))
        break
    
f.close()
```

```python
filenames2 = []
f= open('./test_pistol_yolov5/test2.txt', 'r')
while True:
    line2 = f.readline()
    line2 = line2[0:-1]
    filenames2.append((line2))
    if not line2: 
        print(len(filenames2))
        break
    
f.close()
```

```python
filenames3 = []
f= open('./test_pistol_Kalman3/test.txt', 'r')
while True:
    line3 = f.readline()
    line3 = line3[0:-1]
    filenames3.append((line3))
    if not line3: 
        print(len(filenames3))
        break
    
f.close()
```

```python
with open('./test_pistol_Kalman2/merge.txt', 'w') as outfile:
    for filename in filenames:
        if filename == '':
            break
        with open(filename) as file:
            for line in file:
                outfile.write(line)
```

```python
with open('./test_pistol_yolov5/merge2.txt', 'w') as outfile:
    for filename2 in filenames2:
        if filename2 == '':
            break
        with open(filename2) as file:
            for line2 in file:
                outfile.write(line2)
```

```python
with open('./test_pistol_Kalman3/merge2.txt', 'w') as outfile:
    for filename3 in filenames3:
        if filename3 == '':
            break
        with open(filename3) as file:
            for line3 in file:
                outfile.write(line3)
```

```python
df = pd.read_csv(
    './test_pistol_Kalman2/merge.txt', sep=" ",header=None)

df2 = pd.read_csv(
    './test_pistol_yolov5/merge2.txt', sep=" ",header=None)

df3 = pd.read_csv(
    './test_pistol_Kalman3/merge2.txt', sep=" ",header=None)
```
